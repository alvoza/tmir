```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('paris.jpg')
```

```python
problema_uno(img,30)
```

```python
def problema_uno(img, num):
    
    ### Dimensiones de la imagen.
    alto = img.shape[0]
    ancho = img.shape[1]
    
    imgSalida = np.zeros((alto, ancho, 3))
    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            if(num==1):
                imgSalida[i, j] = [0, 0, blue]
            elif(num==2):
                imgSalida[i, j] = [0, green, 0]
            elif(num==3):
                imgSalida[i, j] = [red, 0, 0]
            elif(num==10):
                imgSalida[i, j] = [red, green, 0]
            elif(num==20):
                imgSalida[i, j] = [0, green, blue]
            elif(num==30):
                imgSalida[i, j] = [red, 0, blue]
            
    cv2.imwrite("Imgane_Salida.jpg", imgSalida)
    imgTemp = cv2.imread("Imgane_Salida.jpg")
    #imgTemp = cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB)
    plt.imshow(imgTemp)
```
