```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('paris.jpg')
```

```python
problema_cuatro(img)
```

```python
def problema_cuatro(img):
    alto = img.shape[0]
    ancho = img.shape[1]
    
    r = img[:,:,0].reshape((alto*ancho, 1))
    g = img[:,:,1].reshape((alto*ancho, 1))
    b = img[:,:,2].reshape((alto*ancho, 1))
    promedio = r*g*b/3
    
    plt.hist(x=r, bins='auto', color='red',
                            alpha=0.7, rwidth=0.85)
    plt.show()
    
    plt.hist(x=g, bins='auto', color='green',
                            alpha=0.7, rwidth=0.85)
    plt.show()
    
    plt.hist(x=b, bins='auto', color='blue',
                            alpha=0.7, rwidth=0.85)
    plt.show()
    
    plt.hist(x=promedio, bins='auto', color='gray',
                            alpha=0.7, rwidth=0.85)
    plt.show()
```
