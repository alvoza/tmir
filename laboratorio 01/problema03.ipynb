```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('tikal.jpg')
```

```python
def problema_tres(img):
    ### Dimensiones de la imagen.
    alto = img.shape[0]
    ancho = img.shape[1]
    
    imgSalida1 = np.zeros((alto, ancho, 3))
    imgSalida2 = np.zeros((alto, ancho, 3))
    imgSalida3 = np.zeros((alto, ancho, 3))
    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            imgSalida1[i, j] = [red, red, red]
            imgSalida2[i, j] = [green, green, green]
            imgSalida3[i, j] = [blue, blue, blue]
            
    cv2.imwrite("Imagen_Salida_red.jpg", imgSalida1)
    imgTemp = cv2.imread("Imagen_Salida_red.jpg")
    plt.imshow(imgTemp)
    cv2.imwrite("Imagen_Salida_green.jpg", imgSalida2)
    imgTemp2 = cv2.imread("Imagen_Salida_green.jpg")
    plt.imshow(imgTemp2)
    cv2.imwrite("Imagen_Salida_blue.jpg", imgSalida3)
    imgTemp3 = cv2.imread("Imagen_Salida_blue.jpg")
    plt.imshow(imgTemp3)
```

```python
problema_tres(im)
```
