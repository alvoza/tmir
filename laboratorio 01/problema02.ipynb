```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ima1 = cv2.imread('./imagen1/imagen1_salida_gray_rojo.jpg')
ima2 = cv2.imread('./imagen1/imagen1_salida_gray_verde.jpg')
ima3 = cv2.imread('./imagen1/imagen1_salida_gray_azul.jpg')
ima1.shape
```

```python
ima1[:,:,1]
```

```python
def problema_dos(img1, img2, img3):
    
    alto = img1.shape[0]
    ancho = img1.shape[1]
    imgSalida = np.zeros((alto, ancho, 3))
    
    imgSalida[:,:,0] = ima1[:,:,0]
    imgSalida[:,:,1] = ima2[:,:,0]
    imgSalida[:,:,2] = ima3[:,:,0]
    
    cv2.imwrite("Imgane_Salida.jpg", imgSalida)
    imgTemp = cv2.imread("Imgane_Salida.jpg")
    #imgTemp = cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB)
    plt.imshow(imgTemp)
```

```python
problema_dos(ima1, ima2, ima3)
```
