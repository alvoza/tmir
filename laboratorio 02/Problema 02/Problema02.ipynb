```python
def binarizar(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    th, im_th_tz = cv2.threshold(gray, thresh=120, maxval=255, type=cv2.THRESH_BINARY)
    plt.imshow(im_th_tz, cmap="gray")
    cv2.imwrite('opencv_binarizada2.jpg', im_th_tz)
    return im_th_tz
```

```python
def suma2(image1, image2):
    alto, ancho,c = image1.shape
    imgSalida = np.zeros((alto, ancho, 3))
    for i in range(0, alto):
        for j in range(0, ancho):
            p = image1[i, j]
            p2 = image2[i, j]
            b = p[2]
            g = p[1]
            r = p[0]
            b2 = p2[2]
            g2 = p2[1]
            r2 = p2[0]
            imgSalida[i, j] = [r+r2, g+g2, b+b2]
    
    cv2.imwrite("Imgane_Salida.jpg", imgSalida)
    imgTemp = cv2.imread("Imgane_Salida.jpg")
    imgTemp = cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB)
    plt.imshow(imgTemp)
```

```python
def o_and(image1, image2):
    alto, ancho = image1.shape
    imgSalida = np.zeros((alto, ancho))
    for i in range(0, alto):
        for j in range(0, ancho):
            p = image1[i, j]
            p2 = image2[i, j]
            imgSalida[i, j] = p and p2
    
    cv2.imwrite("Imgane_and.jpg", imgSalida)
    plt.imshow(imgSalida,cmap="gray" )
```
