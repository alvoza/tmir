```python
### Dimensiones de la imagen.
alto = img.shape[0]
ancho = img.shape[1]
```

```python
imgSalida = np.zeros((alto, ancho, 3)) #Arreglo de salida para RGB
```

### Escala de Grises

```python
gray_green = np.zeros((alto, ancho, 1))
gray_blue = np.zeros((alto, ancho, 1))
gray_red = np.zeros((alto, ancho, 1))

gray_aritmetica = np.zeros((alto, ancho, 1))
gray_ponderado = np.zeros((alto, ancho, 1))
```

```python
for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i,j]
        gray_blue[i,j] = pixel[2]
        gray_green[i,j] = pixel[1]
        gray_red[i,j] = pixel[0]
        
        gray_aritmetica[i,j] = np.mean(pixel)
        gray_ponderado[i,j] = int(0.299*pixel[2] + 0.587*pixel[1] + 0.114*pixel[0])
```

```python
def write_and_show(strImg, arrImg):
    cv2.imwrite(strImg, arrImg)
    imgTemp = cv2.imread(strImg)
    imgTemp = cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB)
    plt.imshow(imgTemp)
```

```python
write_and_show("gray_rojo.jpg", gray_red)
```

```python
write_and_show("gray_verde.jpg", gray_green)
```

```python
write_and_show("gray_azul.jpg", gray_blue)
```

```python
def problema_uno(img, num):
    
    ### Dimensiones de la imagen.
    alto = img.shape[0]
    ancho = img.shape[1]
    
    imgSalida = np.zeros((alto, ancho, 3))
    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            if(num==1):
                imgSalida[i, j] = [0, 0, blue]
            elif(num==2):
                imgSalida[i, j] = [0, green, 0]
            elif(num==3):
                imgSalida[i, j] = [red, 0, 0]
            elif(num==10):
                imgSalida[i, j] = [red, green, 0]
            elif(num==20):
                imgSalida[i, j] = [0, green, blue]
            elif(num==30):
                imgSalida[i, j] = [red, 0, blue]
            
    cv2.imwrite("Imgane_Salida.jpg", imgSalida)
    imgTemp = cv2.imread("Imgane_Salida.jpg")
    #imgTemp = cv2.cvtColor(imgTemp, cv2.COLOR_BGR2RGB)
    plt.imshow(imgTemp)
```

```python
def problema_tres(img):
    ### Dimensiones de la imagen.
    alto = img.shape[0]
    ancho = img.shape[1]
    
    imgSalida1 = np.zeros((alto, ancho, 3))
    imgSalida2 = np.zeros((alto, ancho, 3))
    imgSalida3 = np.zeros((alto, ancho, 3))
    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            imgSalida1[i, j] = [red, red, red]
            imgSalida2[i, j] = [green, green, green]
            imgSalida3[i, j] = [blue, blue, blue]
            
    cv2.imwrite("Imagen_Salida_red.jpg", imgSalida1)
    imgTemp = cv2.imread("Imagen_Salida_red.jpg")
    plt.imshow(imgTemp)
    cv2.imwrite("Imagen_Salida_green.jpg", imgSalida2)
    imgTemp2 = cv2.imread("Imagen_Salida_green.jpg")
    plt.imshow(imgTemp2)
    cv2.imwrite("Imagen_Salida_blue.jpg", imgSalida3)
    imgTemp3 = cv2.imread("Imagen_Salida_blue.jpg")
    plt.imshow(imgTemp3)
```

```python
def problema_cinco(img):
    alto = img.shape[0]
    ancho = img.shape[1]
    gray_ponderado = np.zeros((alto, ancho, 1))
    
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i,j]
            gray_blue[i,j] = pixel[2]
            gray_green[i,j] = pixel[1]
            gray_red[i,j] = pixel[0]

            gray_ponderado[i,j] = int(0.299*pixel[2] + 0.587*pixel[1] + 0.114*pixel[0])
    
    cv2.imwrite("Imagen_ponderada.jpg", gray_ponderado)
    imgTemp = cv2.imread("Imagen_ponderada.jpg")
    plt.imshow(imgTemp)
```

## Problema 04

```python
def problema_cuatro(img):
    alto = img.shape[0]
    ancho = img.shape[1]
    
    r = img[:,:,0].reshape((alto*ancho, 1))
    g = img[:,:,1].reshape((alto*ancho, 1))
    b = img[:,:,2].reshape((alto*ancho, 1))
    promedio = r*g*b/3
    
    plt.hist(x=r, bins='auto', color='red',
                            alpha=0.7, rwidth=0.85)
    plt.show()
    
    plt.hist(x=g, bins='auto', color='green',
                            alpha=0.7, rwidth=0.85)
    plt.show()
    
    plt.hist(x=b, bins='auto', color='blue',
                            alpha=0.7, rwidth=0.85)
    plt.show()
    
    plt.hist(x=promedio, bins='auto', color='gray',
                            alpha=0.7, rwidth=0.85)
    plt.show()
```
