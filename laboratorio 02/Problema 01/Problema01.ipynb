```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_str = "img1.jpg"
img = cv2.imread(img_str)
binarizar(img)
```

```python
def binarizar(image):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    th, im_th_tz = cv2.threshold(gray, thresh=120, maxval=255, type=cv2.THRESH_BINARY)
    plt.imshow(im_th_tz, cmap="gray")
    cv2.imwrite('opencv_binarizada2.jpg', im_th_tz)
    return im_th_tz
```
